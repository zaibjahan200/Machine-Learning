```python
question_cols = [col for col in df.columns if '_q' in col]

df[question_cols] = df[question_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
ocd_dimensions = {
    'Contamination_and_Washing': ['contamination_q1', 'contamination_q2', 'contamination_q3', 'contamination_q4', 'contamination_q5'],
    'Checking_Behavior': ['checking_q1', 'checking_q2', 'checking_q3', 'checking_q4', 'checking_q5'],
    'Ordering/Symmetry': ['ordering_q1', 'ordering_q2', 'ordering_q3', 'ordering_q4', 'ordering_q5'],
    'Hoarding/Collecting': ['hoarding_q1', 'hoarding_q2', 'hoarding_q3', 'hoarding_q4'],
    'Intrusive_Thoughts': ['intrusive_thoughts_q1', 'intrusive_thoughts_q2', 'intrusive_thoughts_q3', 'intrusive_thoughts_q4', 'intrusive_thoughts_q5'],
    'Mental_Compulsions_and_Rituals': ['mental_compulsions_q1', 'mental_compulsions_q2', 'mental_compulsions_q3', 'mental_compulsions_q4'],
    'Avoidance_Behavior': ['avoidance_behavior_q1', 'avoidance_behavior_q2', 'avoidance_behavior_q3'],
    'Emotional_Awareness_and_Insights': ['emotional_awareness_q1', 'emotional_awareness_q2', 'emotional_awareness_q3'],
    'Functioning_Behavior': ['functioning_q1', 'functioning_q2', 'functioning_q3', 'functioning_q4', 'functioning_q5']
}
for dim, questions in ocd_dimensions.items():
    df[f'{dim}'] = df[questions].mean(axis=1)
```

```python
question_cols = [col for col in df.columns if '_q' in col]
df.drop(columns=question_cols, inplace=True)
```

```python
df['ocd_overall_score'] = df[[f"{dim}" for dim in ocd_dimensions]].mean(axis=1)
```

```python
def classify_ocd(score):
    if score < 1.5:
        return 0  # Low/No OCD
    elif score < 2.8:
        return 1  # Moderate Risk
    else:
        return 2  # High Risk

df['has_ocd'] = df['ocd_overall_score'].apply(classify_ocd)
```

```python
# Step 1: Convert all to lowercase and strip whitespaces
df['Country or Region'] = df['Country or Region'].str.lower().str.strip()

# Step 2: Define patterns that indicate "Pakistan"
pakistan_keywords = ['pakistan', 'pk', 'punjab pakistan', 'islamabad', 'islam pakistan', 'pak','islam']

# Step 3: Replace all these with 'Pakistan'
df['Country or Region'] = df['Country or Region'].apply(
    lambda x: 'Pakistan' if any(keyword in x for keyword in pakistan_keywords) else x.title()
)
```

```python
import re

def normalize_text(e):
    e = str(e).strip().lower()
    e = re.sub(r'[^\x00-\x7F]+','', e)  # remove non-ASCII
    e = re.sub(r'\s+', ' ', e)         # collapse multiple spaces
    return e

def clean_field(entry):
    e = normalize_text(entry)
    for key, val in replacements.items():
        if key in e:
            return val
    return e.title()

df['Occupation / Field of Study'] = df['Occupation / Field of Study'].apply(clean_field)
```
